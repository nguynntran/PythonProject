# Conditional operators (if, elif, else)

Conditional operators in Python look like this:

```python
if condition:
    true_action()
else:
    false_action()
```

Let's consider an example of checking the age for driving a car:

````
question = 'Can I drive a car?'
print(question)

age = input("Enter your age: ")
age = int(age)

if age >= 18:
    answer = 'Yes, you can drive'
else:
    answer = "No, you can't drive"

print(answer)
````

**Note** - unlike C/C++, in Python there are no brackets for framing commands, for this indentation is used. According to PEP-8, 1 indent is 4 spaces.

The `else` part of the conditional operator is optional:

````
question = 'Can I drive a car?'
print(question)

age = input("Enter your age: ")
age = int(age)

answer = "No, you can't drive"
if age >= 18:
    answer = 'Yes, you can drive'

print(answer)
````

There can be multiple checks:

````
question = 'Can I drive a car?'

age = int(input("Enter your age: "))
country = input("Enter your country (RU/US): ")

if age >= 16 and country == 'US':
    answer = 'Yes, you can drive'
else:
    if age >= 18 and country == 'Russia':
        answer = 'Yes, you can drive'
    else:
        answer = "No, you can't drive"

print(answer)
````

The expression:

```python
else:
    if condition:
```

can be rewritten as::

```python
elif condition:
```

## Ternary conditional operator

We often want to assign one or the other value to a variable depending on the truth of a condition using a conditional operator. In such cases, it is convenient to use the ternary operator:

```python
[if_true] if [expression] else [if_false]
```

````
age = 20
result = 'adult' if age >= 18 else 'child'
result
````

````
age = 16
'adult' if age >= 18 else 'child'
````

Ternary operators can also be used in expressions:

````
x, y = 3, 5

z = 3 + x if x > y else y
print(z)

z = 3 + (x if x > y else y) # Be more careful with the order of calculations
print(z)
````

Conditions can be used consecutively in ternary operators (but it's better not to do so, as it reduces readability).

````
age = 15
'kid' if age < 13 else 'teenager' if age < 18 else 'adult'
````

The ternary operator can be represented as an equivalent expression:

````
nice = True

personality = "nice" if nice else "mean"
print("The cat is:", personality)

personality = ("mean", "nice")[nice]
print("The cat is:", personality)
````

## `dict`

**Dict** (dictionary, associative array, hash table) - a *key-value* store, i.e., it can store pairs of "key-value" objects and retrieve the value of an object by its key. Any hashable object can be used as a key.

It is analogous to `map` from C++, but can work with multiple data types within a single dictionary.

The main operations on a dictionary:
- add a new pair
- get a pair by key
- remove a pair by key

The operation in the hash table starts with calculating the hash function from the key. The resulting hash value $i=hash(key)$ - the index in the array $H$.

### Hash function

**Hash function** - a special function that algorithmically transforms input data of arbitrary length into an output value of fixed length (a bit string).

The algorithm creates a unique hash for each message. If you pass the same message through the algorithm, the output hash will remain unchanged. But if you replace even one letter in the source data, the hash will change beyond recognition:


image.png

image.png

A key property of a hash function - **irreversibility**. You can't just reverse the algorithm and read the original message. Therefore, if a database with hashed passwords falls into the hands of a hacker, he can only crack them by brute force - and this can take several months and even years.

Another important property is the **efficiency of storing** and quickly accessing large amounts of data, as searching by hash code takes much less time than ordinary linear search.

**Applications:**

- Passwords and digital signatures
- File integrity checking
 - For example, when starting a game - to check that the user hasn't boosted cheat codes
 - Version control systems (Git) — to check the integrity of objects in the repository: files, commits, trees, etc.

- Blockchain
 - To guarantee the authenticity of transactions. Each block of the chain adds a hash of the previous operation - when the data changes, the hashes will not match and the network will not accept the changes.




#### Collisions

In general, there is no unambiguous correspondence between the hash code and the original data. After all, we have an infinite number of messages at the input and a strictly fixed string at the output. This is called:

**Сollision** - a case where the hash function gives one code for different input data.

A "good" hash function should satisfy two properties:

- fast calculation
- minimum number of collisions

Sometimes it is possible to avoid collisions altogether. For example, if all the keys of the elements are known in advance, then a certain perfect hash function can be found for them, which will distribute them over the cells of the hash table without collisions.

However, collisions in hash tables are not so rare. Therefore, the collision resolution mechanism is an important component of any hash table.

### Creating Dictionaries

You can create a dictionary in Python either using curly braces or using the `dict` function:

The most convenient way to create a dictionary from two lists (a list of keys and a list of values) is to use the `zip` function:

````
keys = ['ivan', 'peter', -10, (1, -10)]
values = [100, 200, 'new', True]

dict_example = dict(zip(keys, values))
print(dict_example)
````

### Dictionary keys

The key of a dictionary can be any `hashable` object. The definition of hashable from the Python documentation ([link](https://docs.python.org/3/glossary.html#term-hashable)):

>An object is hashable if it has a hash value which **never changes during its lifetime** (it needs a `__hash__()` method), and **can be compared to other objects** (it needs an __eq__() method). Hashable objects which are "equal" to each other must have the same hash value.



You can view the hash value in Python using the `hash` function. Examples of hash values for different types:

````
print(hash(343))
print(hash(-343))
print(hash(True))
print(hash('hello'))
print(hash((1, 2, 3)))
````

Let's check if a list, tuple, and dictionary are hashable:

````
print("list:", [1, 2].__hash__ is not None)
print("tuple:", (1, 2).__hash__ is not None)
print("dict:", {1: 2}.__hash__ is not None)
````

#### Hashability of tuples

**Attention №1** - not all tuples are hashable! A tuple is hashable if all its elements are also hashable:

````
print(hash((1, 2, (3, 4))))
print(hash((1, 2, [3, 4])))
````

#### Hashing `float`

**Attention** - you need to be very careful with hashing `float` and it's better not to hash them at all:

````
hash1 = hash(6.5)
hash2 = hash(6.500000000000001)
hash3 = hash(6.5000000000000001)
hash4 = hash(round(6.50443,2))  # or hash like this

print("Hash 1: ", hash1)
print("Hash 2: ", hash2)
print("Hash 3: ", hash3)
print("Hash 4: ", hash4)
````

**Note** - after restarting the interpreter, complex objects (e.g., strings) will have a different hash function value.

### Access Dictionary Items

To access dictionary elements, you need to pass the key in square brackets:

````
dict_example = {'ivan': 100, 'peter': 200, -10: 'new', (1, -10): True}

print(dict_example['ivan'])
````

Changing, adding, and removing elements also occurs by keys:

````
dict_example = {'ivan': 100, 'peter': 200, -10: 'new', (1, -10): True}

# Adding
dict_example['ivan'] = 200
print(dict_example)

# Changing
dict_example['ivan_the_great'] = 300
print(dict_example)

# Removing
del dict_example['ivan']
print(dict_example)

# Getting by a non-existent key
print(dict_example['ivan_the_greatest'])
````

### Dictionary methods

Let's consider some useful dictionary methods, the full list of methods with descriptions - [link](https://python-reference.readthedocs.io/en/latest/docs/dict/).

To safely retrieve a value from a dictionary, we use the `get` and `setdefault` methods:

````
dict_example = {'ivan': 100, 'peter': 200, -10: 'new', (1, -10): True, 'oleg': 42}
print(dict_example)
print()

# get method - if there is no key, we will get the default value
key = 'oleg'
res = dict_example.get(key, "default_value")
print(res)
print(dict_example)
print()

# setdefault method - if there is no key, we will get the default value and add it to the dictionary
key = 'oleg'
res = dict_example.setdefault(key, "default_value")
print(res)
print(dict_example)
````

We can get all the keys and values from the dictionary using the `keys`, `values`, and `items` methods:

````
print(dict_example)

# Additionally, you need to convert to a list!
print(list(dict_example.keys()))
print(list(dict_example.values()))
print(list(dict_example.items()))
````

We can merge two dictionaries into one using the `update` method:

````
dict_1 = {1: 2, 2: 3, 3: 4}
dict_2 = {3: 5, 4: 6, 5: 7}

print(dict_1)
dict_1.update(dict_2)

print(dict_1)
````

## set

**Set** - a container that allows storing a collection of unique elements.

It is analogous to `unordered_set` from C++, but can contain objects of different types.

Main operations:
- add an element to the set
- remove an element from the set
- check for the presence of an element in the set

The principle of operation is also based on a hash table.

Where it is convenient to use sets:
- counting unique elements in a list
- fast membership checking

### Creating a set

To create a set, use curly braces or the `set` function:

````
set_example = {1, 2, 3}
print(set_example)
print(type(set_example))
````

````
set_example = set([2, 3, 4])
print(set_example)
print(type(set_example))
````

**Note** - an empty set is created only as `set()`, using `{}` will create an empty dictionary.

### Basic set methods

Let's consider some basic methods of sets.

Using the `update` method, we can merge two sets:

````
a = {1, 2, 3}
b = set([2, 3, 4])

b.update(a) # merge one set with another set.
print(a, b)
````

In addition to checking for membership in a set (`in`), there are checks for one set being a subset of another:

````
print(a, b)
print(b.issubset(a))    # is b a subset of a?
print(b.issuperset(a))  # is b a superset of a?
print(b.isdisjoint(a))  # are they disjoint?
````

set operations:

````
a = {1, 2, 3, 5}
b = set([2, 3, 4, 5, 6])

print(a - b)  # all elements in a that are not in b
print(b - a)  # all elements in b that are not in a
print(a | b)  # union of sets
print(a & b)  # intersection
print(a ^ b)  # all elements that exist only in one of the two sets
````

These operations can also be used as methods of objects:

````
print(a.difference(b))             # a - b
print(a.union(b))                  # a | b
print(a.intersection(b))           # a & b
print(a.symmetric_difference(b))   # a ^ b

a.difference_update(b)            # a -= b
a.update(b)                       # a |= b
a.intersection_update(b)          # a &= b
a.symmetric_difference_update(b)  # a ^= b
````

Removing elements from a set is done using the `remove` and `discard` methods:

````
a = {1, 2, 3}
print(a)
a.remove(3)
print(a)
a.remove(3)
print(a)
````

````
a = {1,2,3}
a.discard(3)
a.discard(3)
a
````

### `frozenset`

Since a set is a mutable data type, it cannot be used as a key. However, it has its equivalent tuple - **frozenset**, which differs only in that it is immutable.

````
set_example = {1, 2, 3}
frozenset_example = frozenset(set_example)

print(type(frozenset_example), frozenset_example)
print(type(set_example), set_example)
````

````
some_dict = {}
some_dict[frozenset_example] = True
print(some_dict[frozenset_example])

some_dict[set_example] = False
print(some_dict[set_example])
````
